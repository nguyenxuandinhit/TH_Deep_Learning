```python
# cau8:
rd_array = np.random.randint(0, 21, size=20)
print("Mảng gốc:", rd_array)


condition = (rd_array >= 9) & (rd_array <= 15)

modified_array = rd_array.copy()
modified_array[condition] = -modified_array[condition]

print("Mảng sau khi đổi dấu:", modified_array)
```

```python

three_d_array = np.arange(27).reshape(3, 3, 3)
print("Mảng 3x3x3 gốc:")
print(three_d_array)

sum_along_columns = np.sum(three_d_array, axis=2)
print("\nTổng theo mỗi cột (sum across columns, axis=2):")
print(sum_along_columns)

sum_along_rows = np.sum(three_d_array, axis=1)
print("\nTổng theo mỗi dòng (sum across rows, axis=1):")
print(sum_along_rows)
```

```python
# cau18:
sr = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
print("Series gốc:")
print(sr)
new_index = ['a', 'e', 'c', 'f']
sr_reindexed = sr.reindex(new_index)
print("\nSeries sau khi reindex với index mới:")
print(sr_reindexed)
sr_reindexed_fill = sr.reindex(['a', 'x', 'b', 'y'], fill_value=0)
print("\nSeries sau khi reindex với index mới và điền giá trị 0 cho các phần tử thiếu:")
print(sr_reindexed_fill)
```
