```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/huynhhoc/DataAnalystDeepLearning/main/Data/covid19/countries-aggregated_csv.csv"
df = pd.read_csv(url)

df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(['CountryRegion', 'Date'])
print(df.head(10))

cols = ['Confirmed', 'Recovered', 'Deaths']
df[cols] = df.groupby('CountryRegion')[cols].transform(lambda x: x.interpolate().ffill().fillna(0))
df[cols] = df[cols].astype(int)
print("Ket Qua Dien Du Lieu")
print(df.head(10))

```

```python
df['New_Confirmed'] = df.groupby('CountryRegion')['Confirmed'].diff().fillna(0)

def smooth_outliers(group):
    mu, std = group.mean(), group.std()
    is_outlier = (group < 0) | (group > mu + 3 * std)
    rolling = group.rolling(window=7, min_periods=1, center=True).mean()
    group[is_outlier] = rolling[is_outlier]
    return group

df['New_Confirmed_Cleaned'] = df.groupby('CountryRegion')['New_Confirmed'].transform(smooth_outliers)
df['Confirmed_Cleaned'] = df.groupby('CountryRegion')['New_Confirmed_Cleaned'].cumsum().astype(int)
print(df[['Date', 'CountryRegion', 'Confirmed', 'Confirmed_Cleaned', 'New_Confirmed_Cleaned']].head(10))
```

```python
df['Log_Confirmed'] = np.log1p(df['Confirmed_Cleaned'])
print(df[['Date', 'CountryRegion', 'Confirmed_Cleaned', 'Log_Confirmed']].head(10))
df.to_csv('processed_covid_step_by_step.csv', index=False)
```
